```python
from sklearn.datasets import fetch_20newsgroups

# subset='train' (Train)  , remove=('headers', 'footers', 'quotes')  
train_news= fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'), random_state=156)
X_train = train_news.data
y_train = train_news.target
print(type(X_train))

# subset='test' (Test)  , remove=('headers', 'footers', 'quotes')  
test_news= fetch_20newsgroups(subset='test',remove=('headers', 'footers','quotes'),random_state=156)
X_test = test_news.data
y_test = test_news.target
print('   {0} ,    {1}'.format(len(train_news.data) , len(test_news.data)))

#        ,   .( )
```


#### Count      //

** :    fit( ) CountVectorizer      .
   CountVectorizer fit_transform()  fit()   .
    fit()       feature   .

```python
from sklearn.feature_extraction.text import CountVectorizer

# Count Vectorization feature extraction  . 
cnt_vect = CountVectorizer()
cnt_vect.fit(X_train)
X_train_cnt_vect = cnt_vect.transform(X_train)

X_test_cnt_vect = cnt_vect.transform(X_test)
#   fit( ) CountVectorizer    feature extraction  . 


print('  Text CountVectorizer Shape:',X_train_cnt_vect.shape, X_test_cnt_vect.shape)

#       .
#    .
```

##  regression , , 
####    regression  .

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# LogisticRegression  // . 
lr_clf = LogisticRegression()
lr_clf.fit(X_train_cnt_vect , y_train)
pred = lr_clf.predict(X_test_cnt_vect)
print('CountVectorized Logistic Regression    {0:.3f}'.format(accuracy_score(y_test,pred)))
```

#### TF-IDF    //



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization       . 
tfidf_vect = TfidfVectorizer()
tfidf_vect.fit(X_train)
X_train_tfidf_vect = tfidf_vect.transform(X_train)
X_test_tfidf_vect = tfidf_vect.transform(X_test)

# LogisticRegression  // . 
lr_clf = LogisticRegression()
lr_clf.fit(X_train_tfidf_vect , y_train)
pred = lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Logistic Regression    {0:.3f}'.format(accuracy_score(y_test ,pred)))
# countvectorize  
```

#### stop words   ngram (1,1) (1,2)   



```python
# stop words   ngram (1,1) (1,2) (bigram)  Feature Vectorization .
tfidf_vect = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_df=300 )
tfidf_vect.fit(X_train)
X_train_tfidf_vect = tfidf_vect.transform(X_train)
X_test_tfidf_vect = tfidf_vect.transform(X_test)

lr_clf = LogisticRegression()
lr_clf.fit(X_train_tfidf_vect , y_train)
pred = lr_clf.predict(X_test_tfidf_vect)
print('TF-IDF Vectorized Logistic Regression    {0:.3f}'.format(accuracy_score(y_test ,pred)))
#   .
```

#### GridSearchCV LogisticRegression C   

```python
from sklearn.model_selection import GridSearchCV

#  C    . CV 3 Fold . (fold 3..?)
params = { 'C':[0.01, 0.1, 1, 5, 10]}
grid_cv_lr = GridSearchCV(lr_clf ,param_grid=params , cv=3 , scoring='accuracy' , verbose=1)
grid_cv_lr.fit(X_train_tfidf_vect , y_train)
print('Logistic Regression best C parameter :',grid_cv_lr.best_params_ )

#  C   grid_cv    . 
pred = grid_cv_lr.predict(X_test_tfidf_vect, max_iter=10 )
print('TF-IDF Vectorized Logistic Regression    {0:.3f}'.format(accuracy_score(y_test ,pred)))
```

###  (Pipeline)   GridSearchCV 




  feature vectorization    ML , pipeline      .

```python
from sklearn.pipeline import Pipeline

# TfidfVectorizer  tfidf_vect , LogisticRegression lr_clf   Pipeline
pipeline = Pipeline([
    ('tfidf_vect', TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_df=300)),
    ('lr_clf', LogisticRegression(C=10))
])

#  TfidfVectorizer fit_transform( ) LogisticRegression fit(), predict( )  . 
# pipeline fit( )  predict( )   Feature Vectorization ML / . 
pipeline.fit(X_train, y_train)
pred = pipeline.predict(X_test)
print('Pipeline  Logistic Regression    {0:.3f}'.format(accuracy_score(y_test ,pred)))
```

```python
# Pipeline gridSearch  
#  30 

from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('tfidf_vect', TfidfVectorizer(stop_words='english')),
    ('lr_clf', LogisticRegression())
])

# Pipeline     (_)2   GridSearchCV  
# /    . . 
params = { 'tfidf_vect__ngram_range': [(1,1), (1,2), (1,3)],
           'tfidf_vect__max_df': [100, 300, 700],
           'lr_clf__C': [1,5,10]
}

# GridSearchCV  Estimator  Pipeline  
grid_cv_pipe = GridSearchCV(pipeline, param_grid=params, cv=3 , scoring='accuracy',verbose=1)
grid_cv_pipe.fit(X_train , y_train)
print(grid_cv_pipe.best_params_ , grid_cv_pipe.best_score_)

pred = grid_cv_pipe.predict(X_test)
print('Pipeline  Logistic Regression    {0:.3f}'.format(accuracy_score(y_test ,pred)))
```
